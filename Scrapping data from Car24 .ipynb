# (i) Used cars data (Ghaziabad)

```python
import selenium
from selenium import webdriver
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions
from selenium.webdriver.common.by import By
from selenium.common.exceptions import ElementNotInteractableException, StaleElementReferenceException, NoSuchElementException
import time
import warnings
warnings.filterwarnings('ignore')

driver = webdriver.Chrome(r"C:/Users/acer/Downloads/Compressed/chromedriver_win32_2/chromedriver.exe")
driver.implicitly_wait(0.5)


# URL
url = 'https://www.cars24.com/buy-used-car/?sort=P&storeCityId=132&pinId=201001'
driver.get(url)
driver.maximize_window()
time.sleep(2)
```

```python
# 1500 time we scroll down by 10000 in order to generate more images on the website
for i in range(1500):
    driver.execute_script("window.scrollBy(0, 10000)")
```

```python
# Make empty lists  
Brand=[]
Owner=[]
kilometers=[]
Transmission=[]
Fuel_type=[]
Model=[]
Price=[]



# Extracting Brand from xpath
try:
    for i in driver.find_elements_by_xpath("//h2[@class='_3FpCg']"):
        Brand.append(i.text)
except NoSuchElementException as e:
    Brand.append("_")
        
# Extracting Name from xpath
try:
    for i in driver.find_elements_by_xpath("//div[@class='_1l4fi']/ul/li[2]"):
        Owner.append(i.text)
except NoSuchElementException as e:
    Owner.append('-')
    
# Extracting colour from xpath
try:
    for i in driver.find_elements_by_xpath("//div[@class='_1l4fi']/ul/li[1]"):
        kilometers.append(i.text)
except NoSuchElementException as e:
    kilometers.append('-')
    
# Extracting colour from xpath
try:
    for i in driver.find_elements_by_xpath("//p[@class='cvakB']/span"):
        Transmission.append(i.text)
except NoSuchElementException as e:
    Transmission.append('-')
    
# Extracting Name from xpath
try:
    for i in driver.find_elements_by_xpath("//div[@class='_1l4fi']/ul/li[3]"):
        Fuel_type.append(i.text)
except NoSuchElementException as e:
    Fuel_type.append('-')
    
# Extracting Name from xpath
try:
    for i in driver.find_elements_by_xpath("//p[@class='cvakB']"):
        Model.append(i.text)
except NoSuchElementException as e:
    Model.append('-')        
        
# Extracting colour from xpath
try:
    for i in driver.find_elements_by_xpath("//div[@class='_19zc5']/div[2]"):
        Price.append(i.text)
except NoSuchElementException as e:
    Price.append('-')
```

```python
# Make a DataFrame
Car=pd.DataFrame({})
Car["Brand"] = Brand
Car["Owner"] = Owner
Car["kilometers"] = kilometers
Car["Fuel"] = Fuel_type
Car["Transmission"] = Transmission
Car["Model"] = Model
Car["Price"] = Price
Car
```
